### Combining Condition

```python
# select all the left backs who are 25 or younger
player[
    (player.position == 'LB') & 
    (player.age <= 25)
]
```

```python
# select all the left backs who are 25 or younger and market value 10+
player[
    (player.position == 'LB') & 
    (player.age <= 25) &
    (player.market_value >= 10)
]
```

```python
# select all the left backs who are 25 or younger and market value 10+ Not from Arsenal and Tottenham
player[
    (player.position == 'LB') & 
    (player.age <= 25) &
    (player.market_value >= 10) &
    ~(player.club.isin(['Arsenal','Tottenham']))
]
```

#### Skill Challenge
    1.They are English(nationality) and
    2.Their market value is more than twice the average market value in the league and 
    3.They either have more than 4000 views or are a new signning but not both

```python
player.head()
```

```python
# First Condition
player_nation=player.nationality.eq('England')
```

```python
# Second Condition
market_value=player.market_value > player.market_value.mean() * 2
```

```python
# Third Condition
pop_and_ns=(player.page_views > 4000) ^ (player.new_signing == 1)
```

```python
player[player_nation & market_value & pop_and_ns]
```
