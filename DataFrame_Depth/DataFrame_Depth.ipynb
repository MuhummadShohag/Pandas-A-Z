### Combining Condition

```python
# select all the left backs who are 25 or younger
player[
    (player.position == 'LB') & 
    (player.age <= 25)
]
```

```python
# select all the left backs who are 25 or younger and market value 10+
player[
    (player.position == 'LB') & 
    (player.age <= 25) &
    (player.market_value >= 10)
]
```

```python
# select all the left backs who are 25 or younger and market value 10+ Not from Arsenal and Tottenham
player[
    (player.position == 'LB') & 
    (player.age <= 25) &
    (player.market_value >= 10) &
    ~(player.club.isin(['Arsenal','Tottenham']))
]
```

#### Selection Terminology Recap

        OPERATOR                                         WHAT IS
   1.player.loc[0:2]=====================================slicing
   2.player.loc[player.age > 37]======================Boolean Masking
   3.player.loc[132, 'name']=======================Basic (label-based) indexing
   4.player.loc[[0, 132], ['name','market_value']]=======fancy indexing
   5.player.lookup([0, 132], ['name','market_value'])====fancy indexing

```python
# fancy indexing
player.loc[[0,132],['name','market_value']]
```

```python
player.lookup([0,132],['name','market_value'])
```

```python
player.lookup([0,132],['market_value','name'])
```
