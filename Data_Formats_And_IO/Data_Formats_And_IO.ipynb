### Reading HTML
    HTML: Hyper Text Markup Language

```python
html = pd.read_html('https://en.wikipedia.org/wiki/List_of_European_countries_by_population',header=1)
```

```python
len(html)
```

```python
html[0]
```

```python
countries = html[0].loc[:,['Country.mw-parser-output .nobold{font-weight:normal}(or territory)','Estimate']]
```

```python
countries.head()
```

```python
countries.rename({'Country.mw-parser-output .nobold{font-weight:normal}(or territory)':'Country'},inplace=True,axis=1)
```

```python
countries.head()
```

```python
countries.replace(r'\[Note \d+](.......)','',regex=True)
```
