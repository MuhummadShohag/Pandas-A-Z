## Inner and Outer Joins

    *Inner Joins:
        1.only the common keys are selected
        2.similar to set intersection
    *Outer Joins:
        1.all keys are selected
        2.similiar to set union

```python
# By default how='inner'
pd.merge(ivies_ds, region_ds, how='inner')
```

```python
# how = 'outer' ==> all keys are selected
pd.merge(ivies_ds, region_ds, how='outer')
```

### Left vs Right

    Left:
        *All left keys are selected
    Right:
        *All right keys are selected

```python
# how='left' ==> all left keys are selected
pd.merge(ivies_ds, region_ds, how='left')
```

```python
# how='right' ==> all right keys are selected
pd.merge(ivies_ds, region_ds, how='right')
```

```python
# same as pd.merge(ivies_ds, region_ds, how='left')
pd.merge(region_ds, ivies_ds,how='right')
```

## Join Cardinalities
    
             1-1
        person <-> DNA
       
       Dual-sided uniques
       *both merge objects contain unique values in the respective key
       
            1-m
       book - pages
       One-sided uniques
       *one of the merge objects contain non-unique values
       *in the resulting pd.merge() the records are repeated M times
       
           M-M
       book - author
       Dual-sided non-uniques
       *both merge objects contain non-unique values
       *in the resulting pd.merge() the records are repeated M*M
       

```python
region_ds[region_ds['School Name'].isin(ivies_ds['School Name'])]
```

```python
# one to one merge ==> merging all of the uniques values
pd.merge(ivies_ds, region_ds, how='inner', on='School Name')
```

```python
# one to many ==> merging two dataset which the records are repeated M times
state_ds['School Name'].is_unique
```

```python
region_ds[region_ds['School Name'].isin(state_ds['School Name'])]
```

```python
# Region dataset has duplicates values
region_ds['School Name'].value_counts()
```

```python
pd.merge(state_ds, region_ds, how='inner', on='School Name')
```

## Merging By Index

    Column joins options:
        *automatic
        *the on parameter
        *left_on, right_on params
        
    Index joins options:
        *left_index, right_index

```python
# Index joins options
iv5 = ivies_ds.set_index('School Name')
iv5
```

```python
reg = region_ds.set_index("School Name") 
```

```python
reg
```

```python
# Both common index values comming to resulting
pd.merge(iv5, reg, left_index=True, right_index=True)
```

#### How about marging by index vs column?
    When we want to merging two datasets, one datasets by index and other datasets by columns, this is perfet choice

```python
pd.merge(iv5, region_ds, left_index=True, right_on='School Name')
```
